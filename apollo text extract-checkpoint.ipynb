```python
import numpy as np

import csv
filename="apollo_data.csv"
with open(filename, 'w') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile)
spz=[]
lpx,lpy,lpz=[],[],[]
with open('pse_a11_1.txt') as f:
    lines = f.readlines()
    for i in range(0,int(len(lines)),7):
        line=lines[i]
        if(len(line)!=0):
           
                if (line[1]=='s' and line[0]=='#' and line[2]=='p' and line[3]=='z'):
                   
                    l=line[6:]
                    for numbers in l.split(','):
                        if (numbers!=' ') :
                            spz.append(float(numbers))
                    for numbers in lines[i+1][6:].split(','):
                        if (numbers!=' '):
                            lpx.append(float(numbers))
                    for numbers in lines[i+2][6:].split(','):
                        if (numbers!=' '):
                            lpy.append(float(numbers))
                    for numbers in lines[i+3][6:].split(','):
                        if (numbers!=' '):
                            lpz.append(float(numbers))
               
                            
                            
                            
import pandas as pd
spz_data=pd.DataFrame({'spz':spz})
lp_data=pd.DataFrame({'lpx':lpx,'lpy':lpy,'lpz':lpz})
spz_data.to_csv('spz_data.csv')
lp_data.to_csv('lp_data.csv')        
        
    
    
    
    
```
